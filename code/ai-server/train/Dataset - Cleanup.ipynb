```python
def replace_professor_names(comment, names_dict):
    """Replaces any professor names found in the comment with a random name."""
    words = comment.split()
    for i, word in enumerate(words):
        if word in names_dict:
            words[i] = get_random_name()
    return " ".join(words)
```

```python
for df in [df_1, df_2]:
    df['processed_comments'] = df['comment'].apply(lambda x: replace_professor_names(x, names_to_check_for_dict))
```
