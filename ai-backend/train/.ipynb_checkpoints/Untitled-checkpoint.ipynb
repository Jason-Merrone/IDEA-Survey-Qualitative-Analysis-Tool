```python
import pandas as pd
from datasets import load_dataset
from datasets import Dataset
```

```python
df = pd.read_csv("dataset/feedback-student.xls", encoding='ISO-8859-1', engine='python')
df.head()
```

```python
# Define a mapping for the "quality" column
quality_mapping = {
    'awful': 0,
    'poor': 0,
    'good': 2,
    'average': 1,
    'excellent': 2,
    'awesome': 2
}
```

```python
# Remove rows with "No Comments" in the comment column
df = df[df['comment'] != 'No Comments']
```

```python
df['sentiment'] = df['quality'].map(quality_mapping)
df.head(5000)
```

```python
# Define approved lists of positive and negative attributes
positive_attributes = [
    "clear", "engaging", "knowledgeable", "supportive", "organized",
    "approachable", "enthusiastic", "empathetic", "resourceful", 
    "inspiring", "reliable", "creative", "patient", "respectful", 
    "motivating", "accessible", "passionate", "unique", "helpful",
    "kind", "interactive", "understanding"
]

negative_attributes = [
    "confusing", "unapproachable", "disorganized", "boring", "unfair",
    "dismissive", "inefficient", "negative", "stubborn", "unreliable", 
    "rigid", "pessimistic", "apathetic", "overbearing", "inflexible",
    "unhelpful", "judgemental"
]
```

```python
test_prompt = (
    "Provide a positive attribute for the following student feedback: "
    "'I loved my professor, he knew a lot about math!' Choose from "
    f"{positive_attributes}, or output 'other' if none are relevant. Please output just the attribute and nothing more."
)

import ollama
response = ollama.chat(model='gemma2:27b', messages=[
  {
    'role': 'user',
    'content': test_prompt,
  },
])
print(response['message']['content'])
```

```python
def validate_attribute(attribute, approved_list):
    print(attribute)
    words = attribute.lower().split()  # Split the sentence into words
    for word in words:
        if word in approved_list:
            print("Success")
            return word
    return "other"
```

```python
def message_model(model, message):
    response = ollama.chat(model=model, messages=[
          {
            'role': 'user',
            'content': message,
          },
        ])
    return response['message']['content']
```

```python
def generate_attributes(comment, sentiment):
    if sentiment == 2:
        prompt = (
            f"<start_of_turn>user\nProvide the positive attribute that best fits the professor for the following student feedback: "
            f"'{comment}'. Choose from {positive_attributes}. Please output just the attribute and nothing more.<end_of_turn>"
        )
        attribute = message_model("gemma2:27b", prompt)
        print(comment + ", " + attribute.lower())
        print()
        return validate_attribute(attribute, positive_attributes), None
    elif sentiment == 0:
        prompt = (
            f"<start_of_turn>user\nProvide the negative attribute that best fits the professor for the following student feedback: "
            f"'{comment}'. Choose from {negative_attributes}. Please output just the attribute and nothing more.<end_of_turn>"
        )
        attribute = message_model("gemma2:27b", prompt)
        print(comment + ", " + attribute.lower())
        print()
        return None, validate_attribute(attribute, negative_attributes)
    else:
        return None, None
```

```python
# Initialize new columns for positive and negative attributes
df['positive-attribute'] = None
df['negative-attribute'] = None

# Iterate through each row in the DataFrame
for idx, row in df.iterrows():
    positive_attr, negative_attr = generate_attributes(row['comment'], row['sentiment'])
    df.at[idx, 'positive-attribute'] = positive_attr
    df.at[idx, 'negative-attribute'] = negative_attr
```
